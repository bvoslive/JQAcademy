## Acessando colunas e linhas com .loc[] e .iloc[]

```python
# .loc[ linhas , colunas], os filtros devem ser iguais aos labels do df
heart.loc[ :3 , :] #index 3 para trás e todas as colunas
heart.loc[2:4, ['age', 'sex']] #index 2 ao 4 e as colunas age e sex
heart.loc[30:35, 'fbs'] #index 30 ao 35 e coluna fbs, perceba que retornou uma series
heart.loc[100:104, ['oldpeak', 'thal']] #index 100 ao 104 e colunas oldpeak e thal
```

```python
# .iloc[ linhas , colunas], os filtros devem ser apenas números
heart.iloc[ :3 , :] #index 3 para trás e todas as colunas
heart.iloc[2:4, 0:2] #index 2 ao 4 e as colunas age e sex
heart.iloc[30:35, 5] #index 30 ao 35 e coluna fbs, perceba que retornou uma series
heart.iloc[100:104, [9, 12]] #index 100 ao 104 e colunas oldpeak e thal
```

**Perceba a igualdade nas saídas dos dois comandos**

## Renomeando colunas de um DF

**Há várias maneiras de se fazer isso, vamos mostrar algumas**

```python
# renomando com .rename(), alteração não acontece "inplace", você deve atribuir a mudança ao df novamente
heart = heart.rename(columns={'age' : 'idade', 'sex' : 'sexo'})
heart.head(3)
```

## Filtrando linhas com filtros booleanos 

```python
heart = pd.read_csv('heart.csv')
suicide = pd.read_csv('master.csv')
crime_boston = pd.read_csv('crime_boston.csv', encoding='latin', low_memory=False)
```

```python
#crie um series booleana com a condição desejada, depois passe como filtro para o df usando colchetes ou .loc[]
filt_1 = crime_boston['OFFENSE_CODE_GROUP'] == 'Disorderly Conduct'

crime_boston[filt_1]
crime_boston.loc[filt_1, :] #todas as linhas que respeitam a condição e todas as colunas
```

**Perceba que as duas saídas são iguais**

```python
# usando mais de um filtro de uma vez
filt_1 = crime_boston['OFFENSE_CODE_GROUP'] == 'Disorderly Conduct'
filt_2 = crime_boston['DAY_OF_WEEK'] == 'Wednesday'

crime_boston.loc[filt_1 & filt_2, ['OFFENSE_CODE_GROUP', 'DAY_OF_WEEK', 'DISTRICT', 'YEAR']]
```

**Filtramos linhas que tinham OFFENSE_CODE_GROUP == 'Disorderly Conduct' e DAY_OF_WEEK == 'Wednesday' e depois filtramos apenas as colunas 'OFFENSE_CODE_GROUP', 'DAY_OF_WEEK', 'DISTRICT' e 'YEAR'**

## Ordenando por colunas ou valores

```python
heart = pd.read_csv('heart.csv')
heart.head(3)
```

```python
# colocando colunas em ordem alfabéticas, ascending=True é o valor default
heart_col_alfa = heart.sort_index(axis=1)
heart_col_alfa.head(3)

# colocando colunas em ordem alfabética "descrescente"
heart_col_nao_alfa = heart.sort_index(axis=1, ascending=False)
heart_col_nao_alfa.head(3)

# colocando index em ordem descrescente
heart_ind_decre = heart.sort_index(axis=0, ascending=False)
heart_ind_decre.head(3)
```

```python
# ordenando por valores
heart.sort_values(by='age')
```

## Usando o groupby()

```python
crime_boston = pd.read_csv('crime_boston.csv', encoding='latin', low_memory=False)
crime_boston.head(3)
```

```python
# normalize=True mostra a frequência relativa 
series_perc = crime_boston.groupby(by=['OFFENSE_CODE_GROUP']).DAY_OF_WEEK.value_counts(normalize=True)
series_perc
```

```python
series_perc.loc[ ['Aggravated Assault', 'Aircraft'] , ['Saturday', 'Sunday']]
```

**.value_counts() funciona muito bem com dados categóricos, vamos ver algumas operaçoes que podemos usar com .groupby()**

```python
heart = pd.read_csv('heart.csv')
heart.head(3)
```

```python
heart.groupby(by=['sex', 'target']).agg({'age' : np.mean, 'oldpeak' : np.mean})
```

```python
heart.groupby(by=['sex', 'target']).mean()
```

## Visualizações rápidas

```python
#import básico 
import matplotlib.pyplot as plt
%matplotlib inline

crime_boston = pd.read_csv('crime_boston.csv', encoding='latin', low_memory=False)
heart = pd.read_csv('heart.csv')
```

**Usando Series**
